```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('BA_AirlineReviews.csv')
```

```python
data.head(30)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X = data['ReviewHeader']    # The review text
y = data['VerifiedReview']
```

```python
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42)
```

```python
# Create and train the Naive Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# Predict labels for the test set
y_pred = model.predict(X_test)

# Display accuracy and classification metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Function to predict whether a new review is fake or genuine
def predict_review(review_text):
    review_vector = vectorizer.transform([review_text])
    prediction = model.predict(review_vector)
    return prediction[0]

# Example: test a custom review
predict_review("The flight was delayed and the service was terrible.")
```
